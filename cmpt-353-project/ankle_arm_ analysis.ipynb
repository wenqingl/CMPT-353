```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
```

```python
ankles = pd.read_csv("./clean_dataset/clean_male_ankle_walk.csv", parse_dates=['time'])
arms = pd.read_csv("./clean_dataset/clean_male_arm_walk.csv", parse_dates=['time'])
```

```python
plt.hist(arms['ax'], alpha=0.5, label='wrist')
plt.hist(ankles['ax'],alpha=0.5, label='ankles')

plt.title('The ax of Wrist vs Ankles')
plt.xlabel('ax m/s^2')
plt.ylabel('Frequency')
plt.legend(loc="upper right")
```

```python
plt.hist(arms['ay'], alpha=0.5, label='wrist')
plt.hist(ankles['ay'],alpha=0.5, label='ankles')

plt.title('The ay of Wrist vs Ankles')
plt.xlabel('ay m/s^2')
plt.ylabel('Frequency')
plt.legend(loc="upper right")
```

```python
plt.hist(arms['az'], alpha=0.5, label='wrist')
plt.hist(ankles['az'],alpha=0.5, label='ankles')

plt.title('The az of Wrist vs Ankles')
plt.xlabel('az m/s^2')
plt.ylabel('Frequency')
plt.legend(loc="upper right")
```

```python
ankles['acceleration'] = np.sqrt((ankles['ax'])**2 + (ankles['ay'])**2 + (ankles['az'])**2)
arms['acceleration'] = np.sqrt((arms['ax'])**2 + (arms['ay'])**2 + (arms['az'])**2)
```

```python
plt.hist(arms['acceleration'], alpha=0.5, label='wrist')
plt.hist(ankles['acceleration'],alpha=0.5, label='ankles')

plt.title('The acceleration of Wrist vs Ankles')
plt.xlabel('acceleration m/s^2')
plt.ylabel('Frequency')
plt.legend(loc="upper right")
```
